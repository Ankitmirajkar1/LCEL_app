```python
## Import necessary libraries

import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough ##Pass raw input (context+ question)

import langsmith
```

```python
load_dotenv()
```

```python
scope = """
Generative AI, or GenAI, refers to powerful AI systems that can create new content — text, images, music, even video — by learning from large sets of existing data. 
It’s like teaching a computer to paint, write, or compose, by showing it many real examples. A fun fact: an AI-generated artwork once sold for US$432,500 — proving that GenAI isn’t just useful, it can shock the art world too! 
GenAI also underpins tools that help write code, draft essays or design images — turning creative power into a few keystrokes. As long as we use it thoughtfully, GenAI can turbo-charge creativity and productivity.
"""
```

```python
prompt = PromptTemplate.from_template("""
You must answer the questions strictly based on the context provided. You must not use any prior knowledge, assumptions, or information not contained in the context.
If the answer is not contained within the context, respond with "I don't know the answer. I can Only answer based on the context provided.
Questions:{question}

""")
```

```python
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
```

```python
## LCEL App Chain

chain = ({"context": RunnablePassthrough(), "question": RunnablePassthrough()} | prompt | llm | StrOutputParser())
```

```python
question1 = "what is the fun fact mentioned in the context?"
```

```python
response = chain.invoke({"context": scope, "question": question1})
```
